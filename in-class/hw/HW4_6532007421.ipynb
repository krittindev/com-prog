1. [8 คะแนน] get_product_from_file(textfile)
  เพื่ออ่านไฟล์ข้อมูลผลิตภัณฑ์และคืน dict ที่เป็นข้อมูลของผลิตภัณฑ์นั้น
  * textfile คือไฟล์ข้อมูลผลิตภัณฑ์อยู่ในรูปแบบ .txt 
  * คืนค่า dict ที่เป็นข้อมูลของผลิตภัณฑ์นั้น โดย key และ value ของ dict จะเป็นค่าก่อนและหลังเครื่องหมาย "=" ในไฟล์ข้อมูล
  * ตัวอย่างด้านล่าง ข้อมูลทุกค่าอยู่ในรูปแบบของสตริง แต่วันที่ผลิต (created_date) จะมีการเปลี่ยนรูปแบบการแสดงผลจากสตริง เช่น "01012566" (มี 8 ตัวอักษรเสมอ) จะถูกใส่เป็น "01/01/2566" ใน dict เช่น
    * get_product_from_file("001.txt")
      * คืนค่า 
```
{'pid': '001', 'created_date': '01/01/2566', 'factory_id': 'CBR1', 
'scan_data': '-,-,+,+,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-'}
```
    * get_product_from_file("008.txt")
      * คืนค่า 
```
{'pid': '008', 'created_date': '12/06/2564', 'factory_id': 'CBR3', 'scan_data': '+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-'}
```
    * get_product_from_file("007.txt")
      * คืนค่า 
```
{'created_date': '11/06/2564', 'pid': '007', 'factory_id': 'CBR3', 'scan_data': '-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,+,-,-,-,-,+,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-'}
```

    * <font color="red">ในการสร้าง dict ระวังช่องว่างก่อนและหลังคำ พยายามให้ลบออกให้หมดก่อนใส่เข้าไปใน dict </font>

2. รหัสผลิตภัณฑ์ที่ใช้ตรวจจะมีไฟล์ผลิตภัณฑ์นั้นเสมอ

4. ตรวจสอบตัวอักษรเล็กใหญ่ในการแสดงผลให้ดี เช่น ขนาด S, M, L จะเป็นตัวใหญ่หมด ส่วน key ของ dict จะเป็นตัวอักษรเล็ก (ตาม txt file ที่อ่านมา) pid หรือ factory_id อาจเป็นตัวเลขหรือตัวอักษรใด ๆ ก็ได้ (ตาม txt file ที่อ่านมา) เป็นต้น

````python
# HW4_STRFILE_DICT (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)
# ดูรายละเอียดและตัวอย่างเพิ่มเติมในคำอธิบายด้านบน

def get_product_from_file(textfile):
  product = {}
  data = open(textfile, 'r')
  for line in data:
    key, value = [e.strip() for e in line.split('=')]
    product[key] = value
  data.close()
  product['created_date'] = product['created_date'][:2] + '/' + product['created_date'][2:4] + '/' + product['created_date'][4:]
  return product


#======================================
def cal_defect_ratio(textfile):
  defect_ratio = 0
  product = get_product_from_file(textfile)
  if 'scan_data' not in product:
    return 'error'
  scan_data = product['scan_data'].split(',')
  defect_ratio = sum([1 for e in scan_data if e == '+']) / len(scan_data)
  return round(defect_ratio,2)


#======================================
def cal_defect_box_ratio(textfile):
  defect_box_ratio = 0
  product = get_product_from_file(textfile)
  if 'scan_data' not in product:
    return 'error'
  scan_data = product['scan_data'].split(',')
  n = int(round(len(scan_data)**(1/3), 0))
  orders = [i for i in range(n**3)]
  order_cube = {
    'width': [ e%n for e in orders ],
    'lenght': [ e%n**2//n for e in orders ],
    'height': [ e//n**2 for e in orders ] 
  }

  value_cube = {
    key: [ [ scan_data[j] == '+' for j in range(n**3) if value[j] == i ] for i in range(n) ] for key, value in order_cube.items()
  }

  any_cube = {
    key: [ any(value[i]) for i in range(n) ] for key, value in value_cube.items()
  }

  border_cube = {
    key: { 'max': get_max_border(value), 'min': get_min_border(value) } for key, value in any_cube.items()
  }

  size_cube = {
    key: value['max'] - value['min'] + 1 for key, value in border_cube.items()
  }

  if '+' not in scan_data: return 0.00

  defect_box_ratio =  size_cube['width'] * size_cube['lenght'] * size_cube['height'] / n**3

  return round(defect_box_ratio,2)


#======================================
def create_prod_summary_file(pids):
  if len(pids) == 0: return
  product_summary = open('product_summary.csv', 'w')
  product_summary.write('pid,created_date,factory_id,defect_ratio,defect_box_ratio\n')
  result = []

  for product_id in pids:
    textfile = product_id + '.txt'
    product = get_product_from_file(textfile)
    defect_ratio = cal_defect_ratio(textfile)
    defect_box_ratio = cal_defect_box_ratio(textfile)
    result.append(
        '{},{},{},{},{}\n'.format(
            product['pid'], 
            product['created_date'],
            product['factory_id'],
            defect_ratio,
            defect_box_ratio
        )
    )
  result.sort()
  product_summary.writelines(result)
  product_summary.close()


#======================================
def create_size_summary_file(pids):
  if len(pids) == 0: return
  size_summary = open('size_summary.csv', 'w')
  size_summary.write('size,#\n')
  sizes = {'S': 0, 'M': 0, 'L': 0}

  for product_id in pids:
    textfile = product_id + '.txt'
    defect_box_ratio = cal_defect_box_ratio(textfile)
    sizes[get_size(defect_box_ratio)] += 1
  size_summary.writelines([
      'S,{}\n'.format(sizes['S']),
      'M,{}\n'.format(sizes['M']),
      'L,{}\n'.format(sizes['L'])
  ])
  size_summary.close()




#======================================
# สามารถเขียนฟังก์ชันที่สร้างเองได้ในบริเวณด้านล่างนี้เท่านั้น 

def get_size(defect_box_ratio):
  if defect_box_ratio < 0.33:
    return 'S'
  if defect_box_ratio < 0.67:
    return 'M'
  return 'L'

def get_max_border(l):
  i = len(l) - 1
  while i > -1:
    if l[i]:
      return i
    i -= 1
  return i

def get_min_border(l):
  i = 0
  while i < len(l):
    if l[i]:
      return i
    i += 1
  return i

print(
    get_product_from_file("001.txt"), 
    get_product_from_file("001.txt") == {
        'pid': '001', 
        'created_date': '01/01/2566', 
        'factory_id': 'CBR1', 
        'scan_data': '-,-,+,+,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-'
    }
)

print(
    get_product_from_file("007.txt"),
    get_product_from_file("007.txt") == {
        'created_date': '11/06/2564', 
        'pid': '007', 
        'factory_id': 'CBR3', 
        'scan_data': '-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,+,-,-,-,-,+,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-,-'
    }
)

print(
    get_product_from_file("008.txt"),
    get_product_from_file("008.txt") == {
        'pid': '008', 
        'created_date': '12/06/2564', 
        'factory_id': 'CBR3', 
        'scan_data': '+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-,+,-'
    }
)

# print(cal_defect_ratio("004.txt"), cal_defect_ratio("004.txt") == 0.04)
# print(cal_defect_ratio("008.txt"), cal_defect_ratio("008.txt") == 0.5)
# print(cal_defect_ratio("009.txt"), cal_defect_ratio("009.txt") == 0.99)

# print(cal_defect_box_ratio("001.txt"), cal_defect_box_ratio("001.txt") == 0.22)
# print(cal_defect_box_ratio("008.txt"), cal_defect_box_ratio("008.txt") == 0.75)
# print(cal_defect_box_ratio("010.txt"), cal_defect_box_ratio("010.txt") == 0.33)

# create_prod_summary_file(["001","ABCD","002","008"])
# create_prod_summary_file(["001","199","003","004","ABCD","002","010","009","008","007","005"])
# create_prod_summary_file([])

# create_size_summary_file(["001","009","008"])
# create_size_summary_file(["001","199","003","004","002","010","009","008","007","005"])
# create_size_summary_file([])
# create_size_summary_file(["001","002","003","004","005","007","008","009","010","199","ABCD"])
````
