# งานของนิสิต 

โปรแกรมในส่วนของ ```main()``` ถูกเตรียมเอาไว้แล้ว โดยจะให้ผู้ใช้ใส่
 

*   ลำดับของนิสิตกลุ่มวิชาแรก ```group1``` เช่น ผู้ใช้อาจใส่ ```A1,A2,A3,A4,A5,A6```
*   ลำดับของนิสิตกลุ่มวิชาที่ 2 ```group2``` เช่น ผู้ใช้อาจใส่ ```B1,B2,B3,B4```
*   จำนวนที่นั่งต่อแถว ```seats_per_row``` ซึ่งจะต้องมีมากกว่า 4 และเป็นจำนวนคู่เท่านั้น เช่น 4, 6, 8, 10
*   ต้องการกลับลำดับหรือไม่ ```is_reverse``` เป็น 0 กรณีไม่กลับลำดับนิสิตแต่ละกลุ่ม และเป็น 1 เมื่อต้องการกลับลำดับนิสิตแต่ละกลุ่ม

ตัวอย่างการเก็บดังนี้ 
``` 
group1 = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6']
group2 = ['B1', 'B2', 'B3', 'B4'] 
seats_per_row = 4
is_reverse = 0 
```

## งานที่ต้องทำ

1. ให้เขียนฟังก์ชัน ```generate_seating_sequence(group1,group2,seats_p_row)``` ที่คืนค่าเป็น ```seating_sequence``` ซึ่งเป็นลำดับที่นั่งของนิสิตสองกลุ่ม รวมถึงกรณีที่ไม่มีคนนั่งด้วย (มีความยาวเท่ากับที่นั่งทั้งหมดที่นั่งที่จุได้ในห้อง) เช่น สำหรับกรณีตัวอยางด้านบน จะได้
``` 
seating_sequence = ['A1', 'B1', 'A2', 'B2', 'A3', 'B3', 'A4', 'B4', 'A5', ' -', 'A6', ' -', ' -', ' -', ' -', ' -']
```

2. ให้เขียนฟังก์ชัน ```group_reverse_order(group)``` โดย ```group``` ที่รับเข้ามาจะเป็นลำดับของนิสิตกลุ่มใดก็ได้ ```group1``` หรือ ```group2``` โดยจะคืนค่า ```reverse_group``` ที่กลับลำดับแล้ว เช่น
```
group = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6']
reverse_group = ['A6', 'A5', 'A4', 'A3', 'A2', 'A1'] 
```

3. เขียนฟังก์ชัน ```display_exam_seating(seating_sequence)``` จะนำ ```seating_sequence``` มาจัดลำดับ 4 แถว ตาม ```seats_p_row``` ที่กำหนด เช่น 4 ที่นั่งต่อแถวแบบไม่กลับลำดับจะแสดงผลได้ดังข้างล่างและคำว่า Exam Room จะต้องวางอยู่ตรงกลางของช่อง (ฟังก์ชันนี้ไม่มีการคืนค่า)

```
-------------
| Exam Room |
-------------
|A1|B1|A2|B2|
-------------
|A3|B3|A4|B4|
-------------
|A5| -|A6| -|
-------------
| -| -| -| -|
-------------
```

**** หากนิสิตเขียนโปรแกรมที่เหลือให้โปรแกรมสมบูรณ์โดยไม่ใช้ if, for, หรือ library อื่นๆ จะได้โบนัส 10% ของคะแนนที่ได้ ****

````python
# ให้เขียนโปรแกรมในบริเวณที่กำหนดในเซลล์นี้เท่านั้น ห้ามลบ/แก้ไข ข้อความที่เตรียมไว้ให้



def generate_seating_sequence(group1,group2,seats_p_row):
  seating_sequence = [' -'] * 4 * seats_p_row
  seating_sequence[:len(group1)*2:2] = group1
  seating_sequence[1:len(group2)*2:2] = group2
  return seating_sequence

#------------------------------------------------------------#    
def group_reverse_order(group):
  reverse_group = group[::-1]
  return reverse_group 

#------------------------------------------------------------#    
def display_exam_seating(seating_sequence):
  seats_p_row = len(seating_sequence) // 4
  heading = '|' + ' ' * (seats_p_row - 3) + "Exam Room" + ' ' * (seats_p_row - 3) + '|'
  divider = '-' * (1 + seats_p_row * 3)
  first_row = '|' + '|'.join(seating_sequence[0 * seats_p_row:0 * seats_p_row + seats_p_row]) + '|'
  second_row = '|' + '|'.join(seating_sequence[1 * seats_p_row:1 * seats_p_row + seats_p_row]) + '|'
  third_row = '|' + '|'.join(seating_sequence[2 * seats_p_row:2 * seats_p_row + seats_p_row]) + '|'
  forth_row = '|' + '|'.join(seating_sequence[3 * seats_p_row:3 * seats_p_row + seats_p_row]) + '|'
  print(divider)
  print(heading)
  print(divider)
  print(first_row)
  print(divider)
  print(second_row)
  print(divider)
  print(third_row)
  print(divider)
  print(forth_row)
  print(divider)

#------------------------------------------------------------#    
def main():
    
    group1 = input().split(",") # A1,A2,A3
    group2 = input().split(",") # C1,C2,C3,C4
    seats_p_row = int(input())  # 4, 6, 8, 10
    is_reverse = int(input())  # 0, 1
    
    if is_reverse:
        group1 = group_reverse_order(group1)
        group2 = group_reverse_order(group2)
        
    if len(group1) > 2*seats_p_row or len(group2) > 2*seats_p_row:
        print("Room is too small. Get a bigger exam room!")
        return
    
    if seats_p_row < 4 or seats_p_row % 2 != 0:
        print("seats_p_row has to be more than 4 and has an even number.")
        return
    
    seating_sequence = generate_seating_sequence(group1,group2,seats_p_row)
    display_exam_seating(seating_sequence)

def test(group1, group2, seats_p_row, is_reverse):
  seating_sequence = generate_seating_sequence(group1,group2,seats_p_row)
  display_exam_seating(seating_sequence)

def tests():
  AS = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12']
  BS = ['B1', 'B2', 'B3', 'B4', 'B4', 'B6', 'B7', 'B8', 'B9', 'B10', 'B11', 'B12']
  test(AS[:4], BS[:4], 4, 0)
  test(AS[:4], BS[:6], 4, 0)
  test(AS[:6], BS[:4], 4, 0)
  test(AS[:6], BS[:6], 4, 0)
  test(AS[:6], BS[:6], 6, 0)
  test(AS[:6], BS[:8], 6, 0)
  test(AS[:8], BS[:6], 6, 0)
  test(AS[:8], BS[:8], 6, 0)

# tests()
main()


````
